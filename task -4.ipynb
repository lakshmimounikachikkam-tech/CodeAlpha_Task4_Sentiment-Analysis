```python
import pandas as pd
from textblob import TextBlob
data = {
    "Text": [
        "I love this product",
        "This service is excellent",
        "The product is okay",
        "I am not satisfied",
        "This is the worst experience"
    ]
}
df = pd.DataFrame(data)
def analyze_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
df["Sentiment"] = df["Text"].apply(analyze_sentiment)
print(df)
```

(2). Use NLP techniques and lexicons to detect specific emotions

```python
from textblob import TextBlob
texts = [
    "I am very happy today",
    "I am feeling sad",
    "I am angry with the service",
    "I am scared about the exam",
    "Wow! This is amazing"
]
emotion_lexicon = {
    "happy": "Joy",
    "amazing": "Joy",
    "sad": "Sadness",
    "angry": "Anger",
    "scared": "Fear",
    "wow": "Surprise"
}
def detect_emotion(text):
    text = text.lower()

    for word, emotion in emotion_lexicon.items():
        if word in text:
            return emotion
    return "Neutral"
for text in texts:
    print(text, "->", detect_emotion(text))
```

```python
Objective:To use NLP techniques and lexicons to identify emotions expressed in text.
```

(3). Apply analysis on data from sources like Amazon reviews,social media and news site

```python
import pandas as pd
from textblob import TextBlob
data = {
    'Text': [
        'This product is amazing and works perfectly!', 
        'I am very disappointed with the service.',    
        'The company announced a new product launch today.
        'Had a great experience using this app.',        
        'The latest update caused several issues.'       
    ]
}
df = pd.DataFrame(data)
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
df['Sentiment'] = df['Text'].apply(get_sentiment)
print(df)
print("\nSentiment Counts:")
print(df['Sentiment'].value_counts())
```

```python
objective:To perform sentiment analysis on text collected from Amazon reviews, social media, and news sites.
```

```python
(4). Understanding public opinion and trends through sentiment patterns
```

```python
import matplotlib.pyplot as plt
sentiment_counts = df['Sentiment'].value_counts().reindex(
    ['Positive', 'Negative', 'Neutral'],
    fill_value=0
)
plt.figure(figsize=(6,4))
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(6,6))
plt.pie(
    sentiment_counts.values,
    labels=sentiment_counts.index,
    autopct='%1.1f%%'
)
plt.title("Sentiment Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
sentiment_counts = df['Sentiment'].value_counts().reindex(
    ['Positive', 'Negative', 'Neutral'],
    fill_value=0
)
plt.figure(figsize=(6,4))
plt.plot(sentiment_counts.index, sentiment_counts.values, marker='o')
plt.title('Public Opinion Trend Through Sentiment Patterns')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```

```python
objective:To identify public opinion and emerging trends through sentiment patterns in textual data.
```

(5). Use results to inform marketing,product development,or social insights

```python
positive = (df['Sentiment'] == 'Positive').sum()
negative = (df['Sentiment'] == 'Negative').sum()
neutral = (df['Sentiment'] == 'Neutral').sum()

print("Marketing Insight:")
if positive > negative:
    print("Most customers have positive opinions. Marketing campaigns can highlight strengths and positive feedback.")
else:
    print("Customer satisfaction needs improvement before major marketing efforts.")
print("\nProduct Development Insight:")
if negative > 0:
    print("Negative feedback can be analyzed to identify areas for product improvement.")
else:
    print("Very few negative comments indicate good product performance.")
print("\nSocial Insight:")
print(f"Positive: {positive}, Negative: {negative}, Neutral: {neutral}")
print("Sentiment patterns help understand public opinion and overall trends.")
```
